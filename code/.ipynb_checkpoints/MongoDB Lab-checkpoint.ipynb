## Navigate to wikipedia and pull down a table of your choice. Here is a sample of what I did to pull down the Harrison Ford movies table to help you out.

```python
import urllib2
from bs4 import BeautifulSoup
import json
import pandas as pd

soup = BeautifulSoup(urllib2.urlopen('https://en.wikipedia.org/wiki/List_of_hairstyles').read())
tables = soup.find_all('table', class_="wikitable")
listy = []
for idx, table in enumerate(tables):
    if idx == 0:
        for idx2, row in enumerate(table.find_all('tr')):
            if idx2 > 0:
                datay = row.find_all('td')
                listy += [{'Hairstyle':datay[0].text,'Description':datay[2].text}]                
hair = pd.DataFrame(listy)
hair.head()
```

## Now it's time to make our new mongodb document! Let's insert our wiki table's values into it.

```python
#Creating a new table with the name table_name here
hairstyles = db.hair

#This removes everything from the table. Works on the same principle as find
hairstyles.remove({})

#To insert, just pass in the values one at a time
[hairstyles.insert_one(i) for i in listy]
```

```python
# What is the average number of floors for a building in this list?
import numpy as np
np.mean(df_tall['Floors'])
```

```python
# How many of the builidings are in the United States?
listy = []
for idx, i in enumerate(tallest.find({'Country': {'$regex' : 'States$'}})):
    listy.append(idx+1)
listy[-1]
```

```python
# How many buildings are over 1,000 ft?
listy = []
for idx, i in enumerate(tallest.find({'Height in feet': {'$regex' : '.*[0-9]{1},.*'}})):
    listy.append(idx+1)
listy[-1]
```

```python
# How many hairstyles are longer than 7 letters
listy = []
for idx, i in enumerate(hairstyles.find({'Hairstyle': {'$regex' : '^\w{8}'}})):
    listy.append(idx+1)
listy[-1]
```

```python
# How many hairstyles are 2 or more words?
listy = []
for idx, i in enumerate(hairstyles.find({'Hairstyle': {'$regex' : '^\W*\w+\s{1}\w+\W*$'}})):
    listy.append(idx+1)
listy[-1]
```
